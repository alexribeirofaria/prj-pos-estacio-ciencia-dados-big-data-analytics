```python
from sklearn.datasets import load_iris
import pandas as pd
data_iris = load_iris()
iris_df= pd.DataFrame(data_iris.data, columns= data_iris.feature_names)
iris_df['encoded_target'] = data_iris.target
iris_df
```

```python
# Lidando com Dados Nulos e Corrompidos
iris_df.iloc[0, 0] = None 
iris_df.iloc[0, 1] = None
iris_df.iloc[0, 2] = None
iris_df.iloc[1, 0] = None
iris_df.iloc[4, 3] = None
iris_df.iloc[149, 2] = None
iris_df.iloc[3, 3] = -9999999999999999999999999999
#iris_df.iloc[149, 2] = "cwcwwc"
iris_df
```

```python
# Regularização de Dados OBS.: Dados incompativéis com apresentação da aula

from sklearn.preprocessing import MinMaxScaler
data_iris = load_iris()
iris_df= pd.DataFrame(data_iris.data, columns= data_iris.feature_names)
iris_df['encoded_target'] = data_iris.target
scaler = MinMaxScaler()
print(scaler.fit(iris_df))
novo_iris_df = scaler.transform(iris_df)
novo_iris_df = pd.DataFrame(
    novo_iris_df,
    columns=iris_df.columns
).round(2)
novo_iris_df

```
