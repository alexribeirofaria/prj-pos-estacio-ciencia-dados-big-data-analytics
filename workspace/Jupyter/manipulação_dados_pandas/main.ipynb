```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
def iris_df():
    """
    Junta iris do CSV + sklearn mantendo padrão limpo:
    sepal_length, sepal_width, petal_length, petal_width, species, encoded_target
    """

    # -----------------------
    # 1. CSV
    # -----------------------
    df_csv = pd.read_csv("Iris.csv")

    df_csv = df_csv.rename(columns={
        "SepalLengthCm": "sepal_length",
        "SepalWidthCm": "sepal_width",
        "PetalLengthCm": "petal_length",
        "PetalWidthCm": "petal_width",
    })

    df_csv = df_csv.drop(columns=["Id"])

    df_csv["encoded_target"] = df_csv["Species"].map({
        "Iris-setosa": 0,
        "Iris-versicolor": 1,
        "Iris-virginica": 2,
    })

    # -----------------------
    # 2. sklearn
    # -----------------------
    iris = load_iris()

    df_sk = pd.DataFrame(
        iris.data,
        columns=[
            "sepal_length",
            "sepal_width",
            "petal_length",
            "petal_width"
        ]
    )

    df_sk["encoded_target"] = iris.target

    # -----------------------
    # 3. JOIN SEGURO (por índice)
    # -----------------------
    df = df_csv.copy()
    df["encoded_target_sk"] = df_sk["encoded_target"]

    # -----------------------
    # 4. species limpa
    # -----------------------
    target_names = ["setosa", "versicolor", "virginica"]
    df["species"] = df["encoded_target"].map(lambda x: target_names[int(x)])

    # -----------------------
    # 5. seleção final
    # -----------------------
    df = df[[
        "sepal_length",
        "sepal_width",
        "petal_length",
        "petal_width",
        "species",
        "encoded_target"
    ]]

    return df, target_names
```

```python
# Usando Metodo Merge 
df1 = pd.DataFrame({
    'lkey': ['foo', 'bar', 'baz', 'foo'],
    'value': [1, 2, 3, 5]
})

df2 = pd.DataFrame({
    'rkey': ['foo', 'bar', 'baz', 'foo'],
    'value': [5, 6, 7, 8]
})

print(
    df1.merge(
        df2,
        left_on='lkey',
        right_on='rkey',
        suffixes=('_left', '_right')
    )
)
```

```python
# Por padrão o pandas utiliza left join 
df = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
    'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']
})

other = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'B': ['B0', 'B1', 'B2']
})

resultado = df.join(
    other,
    lsuffix='_caller',
    rsuffix='_other'
)
resultado
```

```python
# Usando Right Join com pandas 

df = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
    'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']
})

other = pd.DataFrame({
    'key': ['K0', 'K1', 'K2'],
    'B': ['B0', 'B1', 'B2']
})

# RIGHT JOIN
resultado = df.merge(other, on='key', how='right')
resultado
```
