```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('../../datasets/heights_weights.csv')
df.sample(5)
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(df['Height'], df['Weight'])

plt.xlabel('Height')
plt.ylabel('Weight')
```

```python
heights_corr = df.corr()
heights_corr
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(heights_corr, annot=True)
```

```python
from sklearn.model_selection import train_test_split

X = df[['Weight']]
Y = df['Height']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.4)
```

```python
x_train.sample()
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression().fit(x_train, y_train)
```

```python
print("Modelo Linear Simples: Intercept = {:.5}".format(linear_model.intercept_))
print("Modelo Linear Simples: Coefficient = {:.5}".format(linear_model.coef_[0]))
print("Modelo Linear Simples: Weight = {:.5} + {:.5}(Height)".format(linear_model.intercept_, linear_model.coef_[0]))
```

```python
print('Training score: ', linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('Testing score: ', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df[['Weight']]
Y = df['Height']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.sample(5)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
print("Modelo Linear Simples: Intercept = {:.5}".format(linear_model.intercept_))
print("Modelo Linear Simples: Coefficient = {:.5}".format(linear_model.coef_[0]))
print("Modelo Linear Simples: Weight = {:.5} + {:.5}(Height)".format(linear_model.intercept_, linear_model.coef_[0]))
```

```python
print('Training score: ', linear_model.score(x_train, y_train))
```

```python
y_pred = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print('Testing score: ', r2_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color='r')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
