```python
import pandas as pd
import plotly.express as px
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.random.randint(100, size=(500,2))
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
kmeans.labels_
```

```python
xdf = pd.DataFrame(X)
```

```python
xdf['class'] = kmeans.labels_
xdf['class'] = xdf['class'].astype('string')
```

```python
xdf.head()
```

```python
px.scatter(xdf,x=0,y=1,color='class', title='K-Means')
```

```python
kmeans.predict([[68, 20], [17, 52], [73,70]])
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=5, min_samples=3).fit(X)
```

```python
xdf_dbscan = pd.DataFrame(X)
```

```python
xdf_dbscan['class'] = clustering.labels_
xdf_dbscan['class'] = xdf_dbscan['class'].astype('string')
def noiser(x):
    if x=='-1':
        return 'NoIsE'
    else:
        return x
xdf_dbscan['class'] =xdf_dbscan['class'].apply(lambda x: noiser(x))
```

```python
px.scatter(xdf_dbscan,x=0,y=1,color='class', title='DBSCAN')
```
